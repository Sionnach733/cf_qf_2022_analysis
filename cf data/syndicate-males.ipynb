predict results of females in lowlands throwdown based on open and qf performances

```python
import numpy as np
import pandas as pd

female_workout_data = pd.read_csv('./merged-males.csv')
female_lowlands_athletes_f = pd.read_csv('./maleSyndicate.csv')
female_workout_data.head(5)
```

```python
female_lowlands_athletes_f.head()
```

```python
merge_result = female_lowlands_athletes_f.merge(female_workout_data, how="inner", on=['id'], suffixes=('', '_delme'))
```

```python
# Discard the columns that acquired a suffix
merge_result = merge_result[[c for c in merge_result.columns if not c.endswith('_delme')]]
```

```python
merge_result['semi-ranking'] = ((merge_result['22.1Rank']*(1-0.39))+(merge_result['22.2Rank']*(1-0.31))+(merge_result['22.3Rank']*(1-0.76)))*0.1+(merge_result['QFOneRank']*(1-0.79))+(merge_result['QFTwoRank']*(1-0.76))+(merge_result['QFThreeRank']*(1-0.62))+(merge_result['QFFourRank']*(1-0.77))+(merge_result['QFFiveRank']*(1-0.84))
```

```python
merge_result[['name', 'semi-ranking']].sort_values(by=['semi-ranking'])
```
