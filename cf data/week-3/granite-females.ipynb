predict results of females in lowlands throwdown based on open and qf performances

```python
import numpy as np
import pandas as pd

athlete_data = pd.read_csv('../merged-females.csv')
semi_roster = pd.read_csv('./femaleGranite.csv')
output_csv = 'granite-females-predictions.csv'
athlete_data.head(5)
```

```python
semi_roster.head()
```

```python
merge_result = semi_roster.merge(athlete_data, how="inner", on=['id'], suffixes=('', '_delme'))
```

```python
# Discard the columns that acquired a suffix
merge_result = merge_result[[c for c in merge_result.columns if not c.endswith('_delme')]]
```

```python
merge_result['semi-ranking'] = ((merge_result['22.1Rank']*(0.46))+(merge_result['22.2Rank']*(0.35))+(merge_result['22.3Rank']*(0.73)))*0.1+(merge_result['QFOneRank']*(0.77))+(merge_result['QFTwoRank']*(0.8))+(merge_result['QFThreeRank']*(0.65))+(merge_result['QFFourRank']*(0.82))+(merge_result['QFFiveRank']*(0.82))
```

```python
merge_result[['name', 'semi-ranking']].sort_values(by=['semi-ranking'])
```

```python
merge_result.to_csv(output_csv, index=False)
```
