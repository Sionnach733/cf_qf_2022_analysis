```python
import numpy as np
import pandas as pd

lowlands_athletes_f = pd.read_csv('./sid-females-predictions.csv')
lowlands_athletes_m = pd.read_csv('./sid-males-predictions.csv')

torian_athletes_f = pd.read_csv('./copa-females-predictions.csv')
torian_athletes_m = pd.read_csv('./copa-males-predictions.csv')

syndicate_athletes_f = pd.read_csv('./atlas-females-predictions.csv')
syndicate_athletes_m = pd.read_csv('./atlas-males-predictions.csv')
```

```python
lowlands_athletes_f = lowlands_athletes_f.sort_values(by=['semi-ranking']).reset_index()
lowlands_athletes_m = lowlands_athletes_m.sort_values(by=['semi-ranking']).reset_index()

torian_athletes_f = torian_athletes_f.sort_values(by=['semi-ranking']).reset_index()
torian_athletes_m = torian_athletes_m.sort_values(by=['semi-ranking']).reset_index()

syndicate_athletes_f = syndicate_athletes_f.sort_values(by=['semi-ranking']).reset_index()
syndicate_athletes_m = syndicate_athletes_m.sort_values(by=['semi-ranking']).reset_index()
```
